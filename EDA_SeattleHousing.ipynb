### Are there any relationships between listing characteristics such as size, amenities, and property type and listing price or review scores?

```python
# Use this, and more code cells, to explore your data. Don't forget to add markdown cells to document your observations and findings.

# select relevant columns and drop rows with missing values
columns = ['listing_id', 'price', 'property_type', 'room_type', 'accommodates', 'bathrooms', 'bedrooms', 'beds', 'amenities', 'review_scores_rating']
listings_df = listings_df[columns].dropna()

# convert categorical variables to numeric using one-hot encoding
listings_df = pd.get_dummies(listings_df, columns=['property_type', 'room_type'], drop_first=True)
```

```python
# calculate the correlation between each feature and the price or review scores
price_correlations = listings_df.corr()['price'].sort_values(ascending=False)
review_score_correlations = listings_df.corr()['review_scores_rating'].sort_values(ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# create a heatmap of the price correlations
plt.figure(figsize=(12, 8))
sns.heatmap(price_correlations.to_frame(), cmap='coolwarm', annot=True, fmt='.2f')
plt.title('Correlation between listing features and price')
plt.show()
```



Looking at the output provided, we can see that the variable 'price' has a perfect positive linear relationship with itself (correlation coefficient of 1), which is expected. The variable 'accommodates' has a strong positive linear relationship with 'price', with a correlation coefficient of 0.648. This suggests that as the number of people a property can accommodate increases, the price of the property tends to increase as well.

Similarly, 'bedrooms' and 'beds' have strong positive linear relationships with 'price', with correlation coefficients of 0.628 and 0.585, respectively. This suggests that as the number of bedrooms and beds in a property increases, the price of the property tends to increase as well.

'Bathrooms' also has a positive linear relationship with 'price', with a correlation coefficient of 0.510, but the relationship is weaker than the relationships observed for 'accommodates', 'bedrooms', and 'beds'. This suggests that the number of bathrooms in a property may not have as much of an impact on the price as the number of bedrooms or beds.

Among the property types, 'Boat' has the strongest positive linear relationship with 'price', with a correlation coefficient of 0.102. This suggests that properties listed as 'Boats' tend to have higher prices than other property types. 'Dorm', 'Other', 'Camper/RV', 'Tent', and 'Bed & Breakfast' have negative linear relationships with 'price', with correlation coefficients ranging from -0.027 to -0.033. This suggests that properties listed as these property types tend to have lower prices than other property types.

The variable 'review_scores_rating' has a weak positive linear relationship with 'price', with a correlation coefficient of 0.055. This suggests that properties with higher review scores tend to have slightly higher prices.

Finally, the variable 'room_type_Shared room' has a strong negative linear relationship with 'price', with a correlation coefficient of -0.159. This suggests that properties listed as shared rooms tend to have lower prices than properties listed as private rooms or entire homes/apartments.

Overall, the correlation coefficients provide useful information on the relationships between the variables in the dataset, and can help guide our analysis and understanding of the factors that influence the price of Airbnb listings in Seattle.

### How does the number of listings and total revenue generated by Airbnb in Seattle change over time?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load the listings dataset
listings_df = pd.read_excel('/content/sample_data/listings.xlsx')
```

```python
# load the calendar dataset
calendar_df = pd.read_excel('/content/sample_data/calendar.xlsx')
```

```python
# Convert date columns to datetime format
calendar_df['date'] = pd.to_datetime(calendar_df['date'])
listings_df['last_review'] = pd.to_datetime(listings_df['last_review'])

# Extract month and year from date columns
calendar_df['month'] = calendar_df['date'].dt.month
calendar_df['year'] = calendar_df['date'].dt.year
listings_df['year'] = listings_df['last_review'].dt.year
```

```python
listings_by_year = listings_df.groupby('year')['id'].count()
listings_by_year
```

```python
# Create a line chart of the number of listings over time
listings_by_year = listings_df.groupby('year')['id'].count()
plt.plot(listings_by_year)
plt.title('Number of Listings by Year')
plt.xlabel('Year')
plt.ylabel('Number of Listings')
plt.show()
```

The data shows the number of listings per year from 2011 to 2016. Here is a statistical summary and analysis of the output:

  - The earliest year in the data is 2011 with only 1 listing, while the latest year is 2016 with 217 listings.
  - The year with the highest number of listings is 2015 with 2894, followed by 2016 with 217.
  - The data is positively skewed, as the majority of the listings are concentrated in the later years (2014-2016), with relatively few listings in the earlier years (2011-2013).
  - There is a significant increase in the number of listings from 2013 to 2014, followed by a much larger increase from 2014 to 2015. However, there is a decrease in the number of listings from 2015 to 2016.

```python
# Create a line chart of the total revenue generated over time
calendar_df['available'] = calendar_df['available'].replace('t', 1).replace('f', 0).astype(int)
calendar_df['revenue'] = calendar_df['price'] * calendar_df['available']
revenue_by_year = calendar_df.groupby('year')['revenue'].sum()
print(revenue_by_year)
plt.plot(revenue_by_year)
plt.title('Total Revenue Generated by Year')
plt.xlabel('Year')
plt.ylabel('Total Revenue (USD)')
plt.show()
```

The large difference between the two years suggests that there was a significant decrease in revenue from 2016 to 2017.

```python
import seaborn as sns

# Create a bar chart of the total revenue generated by year
revenue_by_year = calendar_df.groupby('year')['revenue'].sum().reset_index()
sns.barplot(x='year', y='revenue', data=revenue_by_year, palette='Blues_d')
plt.title('Total Revenue Generated by Year')
plt.xlabel('Year')
plt.ylabel('Total Revenue (USD)')
plt.ticklabel_format(style='plain', axis='y', useOffset=False)
plt.show()
```

 the total revenue generated by Airbnb listings in Seattle in 2016 was approximately 97,708,027 USD while the total revenue generated in 2017 was only 591,428 USD. The large difference between the two years suggests that there was a significant decrease in revenue in 2017.

### Are there any seasonal trends in the Seattle Airbnb market, and if so, how do they affect pricing and availability?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the calendar dataset
calendar_df = pd.read_excel('/content/sample_data/calendar.xlsx')
```

```python
# load the listings dataset
listings_df = pd.read_excel('/content/sample_data/listings.xlsx')
```

```python

# convert date columns to datetime format
calendar_df['date'] = pd.to_datetime(calendar_df['date'])
listings_df['last_review'] = pd.to_datetime(listings_df['last_review'])

# add a day of week column to the calendar dataframe
calendar_df['day_of_week'] = calendar_df['date'].dt.day_name()

# extract month and year from date columns
calendar_df['month'] = calendar_df['date'].dt.month
calendar_df['year'] = calendar_df['date'].dt.year
listings_df['year'] = listings_df['last_review'].dt.year

# clean the 'available' column
calendar_df['available'] = calendar_df['available'].map({'f': 0, 't': 1})
```

```python
# create a line chart of average monthly prices
monthly_prices = calendar_df.groupby(['year', 'month'])['price'].mean()
monthly_prices = monthly_prices.reset_index()
plt.plot(monthly_prices['price'])
plt.title('Average Monthly Prices')
plt.xlabel('Month')
plt.ylabel('Price')
plt.show()
```

 Statistical summary of the data:

  - Mean: 139.834
  - Standard deviation: 9.411
  - Minimum value: 123.818
  - Maximum value: 155.264

Looking at the data, we can see that there is a general increasing trend in the prices over the course of the year, with the exception of a slight dip in the months of September and October. It may be interesting to explore the factors that contribute to this trend, such as seasonal demand or events in the area during certain months. Additionally, the standard deviation of the data is relatively small, indicating that the prices are generally consistent from month to month.

```python
# create a heatmap of listing availability
availability = calendar_df.groupby(['month', 'day_of_week'])['available'].mean().unstack()
sns.heatmap(availability, cmap='YlGnBu')
plt.title('Listing Availability by Day of Week and Month')
plt.xlabel('Day of Week')
plt.ylabel('Month')
plt.show()
```

The heatmap shows the correlation between day of the week and month, with values ranging from 0 to 1. The higher the value, the stronger the correlation between the two variables. The diagonal line of the heatmap represents the correlation between the same day of the week and month, which is always equal to 1.

Looking at the heatmap, we can observe that there is a higher correlation between certain days of the week and months. For example, Fridays in January and Saturdays in June have a higher correlation than other days of the week and months. This could suggest that there are certain patterns in the data that could be explored further.

Overall, the heatmap provides a useful visualization of the correlation between day of the week and month and can be used to identify any patterns or trends in the data.

```python
# create a bar chart of listing types by season
listings_df['season'] = listings_df['last_review'].apply(lambda x: 'winter' if x.month <= 2 or x.month >= 11 else
                                                            'spring' if 3 <= x.month <= 5 else
                                                            'summer' if 6 <= x.month <= 8 else 'fall')
listings_by_season = listings_df.groupby(['season', 'room_type'])['id'].count().unstack()
listings_by_season.plot(kind='bar', stacked=True)
plt.title('Listings by Season and Room Type')
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```

The above figure shows the count of different room types (Entire home/apt, Private room, and Shared room) for each season (fall, spring, summer, and winter).

  - In fall, the most common room type is Entire home/apt with 881 listings, followed by Private room with 471 listings and Shared room with 51 listings.
  - In spring, Entire home/apt and Private room are the most common room types with 39 and 19 listings respectively, and only 2 listings for Shared room.
  - In summer, Entire home/apt is the most common room type with 270 listings, followed by Private room with 98 listings and Shared room with 14 listings.
  - In winter, Entire home/apt is the most common room type with 1351 listings, followed by Private room with 572 listings and Shared room with 50 listings.

We can also see that the highest number of listings is in the winter season, with 1973 total listings, while the lowest number of listings is in the spring season, with only 60 total listings.
